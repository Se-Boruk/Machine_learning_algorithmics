### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
def Function(x):
    y = (1 / 8) * (x**4) + 2*((x+0.7)**3) - 0.5*(x**2) - 38*x - 1
    return y
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
x_list = [x for x in range(-20,21)]
y_list = [Function(x) for x in x_list]
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
import matplotlib.pyplot as plt
plt.plot(x_list,y_list)
plt.title("Wykres funkcji F(x)")
plt.xlabel("Wartosci x")
plt.ylabel("Wartosci y")

#Przybliżenie wykresu
#plt.xlim(2.5 , 4.5)
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
#Wartosc minimalna z y_list
min_y = min(y_list)
#Index wartosci minimalnej w liscie y (taki sam bedzie w liscie x)
arg_x = y_list.index(min_y)
#Wyciagniecie wartosci x
arg_x = x_list[arg_x]
print("Min y:",min_y, "Dla x:", arg_x)

#Wyswietlenie wartosci x i y na wykresie
plt.plot(x_list,y_list)
plt.xlim(arg_x-3, arg_x+3)
plt.ylim(min_y-5, min_y+100)
plt.scatter(arg_x, min_y, c = "red")

plt.title("Wykres funkcji F(x), przyblizenie wartosci minimum")
plt.xlabel("Wartosci x")
plt.ylabel("Wartosci y")
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
import numpy as np
step = 0.2
x_list_step = np.arange(-20,20,step).tolist()
y_list_step = [Function(x) for x in x_list_step]
x_list_step
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
#Wartosc minimalna z y_list
min_y_step = min(y_list_step)
#Index wartosci minimalnej w liscie y (taki sam bedzie w liscie x)
arg_x_step = y_list_step.index(min_y_step)
#Wyciagniecie wartosci x
arg_x_step = x_list_step[arg_x_step]

print("Min y:",min_y_step, "Dla x:", arg_x_step)

#Wyswietlenie wartosci x i y na wykresie
plt.plot(x_list_step,y_list_step)
#plt.xlim(arg_x-3, arg_x+3)
#plt.ylim(min_y-5, min_y+100)
plt.scatter(arg_x_step, min_y_step, c = "red")

plt.title("Wykres funkcji F(x), losowe k wartosci w zakresie [-20,20]; wskazana wartosci minimum")
plt.xlabel("Wartosci x")
plt.ylabel("Wartosci y")
```

### Zad 7:

Wygeneruj $k$ losowych wartości $x$ w przedziale [-20, 20](metoda Monte Carlo), a następnie znajdź minimum funkcji $f(x)$.
Wyświetl wykres funkcji.

```python
k = 100
x_list_random = np.random.uniform(low=-20, high=20, size=k).tolist()
x_list_random.sort()
y_list_random = [Function(x) for x in x_list_random]
```

```python
#Wartosc minimalna z y_list
min_y_random = min(y_list_random)
#Index wartosci minimalnej w liscie y (taki sam bedzie w liscie x)
arg_x_random = y_list_random.index(min_y_random)
#Wyciagniecie wartosci x
arg_x_random = x_list_random[arg_x_random]
print("Min y:",min_y_random, "Dla x:", arg_x_random)

#Wyswietlenie wartosci x i y na wykresie
plt.plot(x_list_random,y_list_random)
#plt.xlim(arg_x-3, arg_x+3)
#plt.ylim(min_y-5, min_y+100)
plt.scatter(arg_x_random, min_y_random, c = "red")

plt.title("Wykres funkcji F(x); wskazana wartosci minimum")
plt.xlabel("Wartosci x")
plt.ylabel("Wartosci y")
```
